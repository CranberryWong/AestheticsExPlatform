```python
# Merge user data

project_path = 'file:///Users/chenwang/Pictures/Analysis/'
userdata_path = 'file:///Users/chenwang/Pictures/Analysis/userdata/'

usertable = pd.DataFrame(pd.read_csv(project_path + 'user.csv'), columns=['name','nation','education','gender','design', 'age'])

usernamelist = usertable['name']

usertable = usertable.replace({"China": 0, "Japan": 1,
                   "Others": 0, "Bachelor": 1, "Master": 2, "Doctor": 3,
                   "Male": 0, "Female": 1,
                   "Little": 0, "Some": 1, "Middle": 2, "Pro": 3
                  })

print(usernamelist)

def mergeAllData():
    frames = []

    for u in usernamelist:
        print(u)
        auserdata = pd.read_csv(userdata_path + u + '.csv')
        auserdata = auserdata[auserdata.wid != 90][auserdata.wid != 117][auserdata.wid != 106]
        auserdata = auserdata[auserdata.wid.map(lambda x:x not in range(150,170))]
        userinfo = usertable.loc[(usertable.name == u)]
        auserdata['value'] = 1
        userinfo['value'] = 1
        fixtable = pd.merge(auserdata, userinfo, on='value')
        del fixtable['value']
        frames.append(fixtable)

    #print(frames)
    
    result = pd.concat(frames)
    #print(result)
    #result = result[result.wid.map(lambda x:x not in range(150,170))]
    result.to_csv(os.path.join(os.path.dirname('.'), "total.csv"), index=False)
    print('finish')
        
mergeAllData()
    
```
